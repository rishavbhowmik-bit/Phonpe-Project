## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1. Copy dataset (safe practice)

df_clean = df.copy()

# 2. Rename columns (simplify names)

df_clean.columns = [
"success",
"code",
"from_date",
"to_date",
"category",
"payment_type",
"transaction_count",
"transaction_amount",
"timestamp"
]

# 3. Check duplicates

print("Duplicate rows:", df_clean.duplicated().sum())

# Remove duplicates

df_clean = df_clean.drop_duplicates()

# 4. Check missing values

print("\nMissing values:\n", df_clean.isnull().sum())

# 5. Convert data types

df_clean["transaction_count"] = pd.to_numeric(df_clean["transaction_count"], errors='coerce')
df_clean["transaction_amount"] = pd.to_numeric(df_clean["transaction_amount"], errors='coerce')

# 6. Convert timestamp to datetime

df_clean["timestamp"] = pd.to_datetime(df_clean["timestamp"], unit='ms', errors='coerce')

# 7. Final dataset info

print("\nCleaned Dataset Info:")
df_clean.info()

# 8. Preview cleaned data

df_clean.head()
````
